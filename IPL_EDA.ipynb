```python
def bar_plot(variable):
    var=df[variable]
    varValue=var.value_counts()
    plt.figure(figsize=(9,3))
    plt.bar(varValue.index,varValue)
    plt.xticks(varValue.index,varValue.index.values)
    plt.ylabel('Frequnecy')
    plt.title(variable)
    plt.show()
    print('{}:\n {}'.format(variable,varValue))
```

```python
def hist_plot(vari):
    var=df[vari]
    varValue=var.value_counts()
    plt.figure(figsize=(9,3))
    plt.bar(varValue.index,varValue)
    plt.xticks(varValue.index,varValue.index.values)
    plt.ylabel('Frequnecy')
    plt.title(vari)
    plt.show()
    print('{}:\n {}'.format(vari,varValue))
```

```python
# Boxplot
```

```python
def box_plot(vari, df):
    plt.figure(figsize=(9, 4))
    sns.boxplot(y=df[vari])
    plt.title(f'Box Plot of {vari}')
    plt.ylabel(vari)
    plt.show()
```

```python
colum_plot=['PLAYER NAME', 'AGE', 'COUNTR','TEAM',
       'PLAYING ROLE', 'T-RUNS', 'T-WKTS', 'ODI-RUNS-', 'ODI-SR-B', 'ODI-WKTS',
       'AUCTION YEAR', 'BASE PRICE', 'SOLD PRICE']
for i in colum_plot:
    if df[i].dtype in ['int64', 'float64']:  # Ensure only numerical columns are plotted
        box_plot(i, df)
```
